```python
import pandas as pd
import os
```

#### Juntar todos os meses numa planilha do ano 2019

```python
dataframe = pd.DataFrame()

files = [file for file in os.listdir('./meses_separados')]

for file in files:
    df = pd.read_csv('./meses_separados/' + file)
    dataframe = pd.concat([dataframe, df])

dataframe.to_csv('Ano2019.csv', index=False)
```

```python
dataframe = pd.read_csv('Ano2019.csv')
dataframe.info()
```

#### Limpar todos os valores NaN

```python
nan = dataframe[dataframe.isna().any(axis=1)]
nan.head()
```

```python
dataframe = dataframe.dropna(how='all')
dataframe.head()
```

#### Limpar default data da base de dados

```python
data_or = dataframe[dataframe['Order Date'].str[0:2] == 'Or']
data_or.head()
```

```python
dataframe = dataframe[dataframe['Order Date'].str[0:2] != 'Or']
dataframe.iloc[500:551]
```

##### Modificar o tipo de dados: de string para inteiro

```python
dataframe['Price Each'] = pd.to_numeric(dataframe['Price Each'])
dataframe['Quantity Ordered'] = pd.to_numeric(dataframe['Quantity Ordered'])

new_dataframe = dataframe
```

#### Criar novas colunas (nao existentes na base de dados) para a visualização de certos dados

```python
#Coluna só com o mês
dataframe['Mes'] = dataframe['Order Date'].str[0:2]
dataframe['Mes'] = dataframe['Mes'].astype('int32')
dataframe
```

```python
#Adicionar coluna de vendas (QUANTIDADE DA ENCOMENDA * PREÇO POR PEÇA)

dataframe['Vendas'] = dataframe['Quantity Ordered'] * dataframe['Price Each']
dataframe
```

```python
# Criar uma coluna com a cidade e o estado em que foi realizada a venda

def pegar_cidade(address):
    return address.split(',')[1]
def pegar_estado(address):
    return address.split(',')[2].split(' ')[1]

dataframe['Cidade'] = dataframe['Purchase Address'].apply( lambda x: f'{pegar_cidade(x)} ({pegar_estado(x)})' )

dataframe
```

#### Pergunta 3: Qual seria o melhor horario para investir em publicidade?

```python
data_time = dataframe
data_time['Order Date'] = pd.to_datetime(data_time['Order Date'])
data_time['Hora'] = data_time['Order Date'].dt.hour
data_time['Min'] = data_time['Order Date'].dt.minute
data_time.drop('Horas', axis=1)
```

```python
ordered_time = pd.DataFrame()


hours = [hour for hour, df in data_time.groupby('Hora')]
order_count = data_time.groupby(['Hora'])['Quantity Ordered'].count()

ordered_time['Hora'] = hours
ordered_time['Quantity Ordered'] = order_count
ordered_time
```

```python
fig,ax = plt.subplots()
ordered_time.plot(x='Hora',y='Quantity Ordered', ax=ax, xlabel='Horas', ylabel='numero de encomendas', grid=True)
ax.set_title('Encomendas e horas das mesmas')

#Config xticks
horas = range(0,24)
ax.set_xticks(horas)
ax.set_xticklabels([f'{hour:02d}:00' for hour in hours], rotation=90, ha='right')

image_stream_path = 'publicidade_analise.png'
plt.savefig(image_stream_path, format='png', bbox_inches='tight')
plt.close(fig)

with pd.ExcelWriter('pub_analise.xlsx', engine='xlsxwriter') as writer:
    ordered_time.to_excel(writer, sheet_name='Publicidade', index=False, startrow=0, startcol=7)

    worksheet = writer.sheets['Publicidade']
    worksheet.title = 'Encomendas e horas das mesmas'

    worksheet.insert_image('K7', image_stream_path, {'x_scale':0.5,'y_scale':0.5})
```

```python
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill

wb = load_workbook('pub_analise.xlsx')
ws = wb['Publicidade']

header_style = Font(color='FFFFFF', bold=True)
fill_column_header = PatternFill(start_color='1F497D', end_color='1F497D', fill_type='solid')
fill_first_column_values = PatternFill(start_color='4F81BD',end_color='4F81BD',fill_type='solid')
fill_other_columns =  PatternFill(start_color='D9E1F2',end_color='D9E1F2',fill_type='solid')

for row in ws.iter_rows(min_row=1, max_row=1):
    for i, header_cell in enumerate(row, start=1):
        cell_value = header_cell.value
        if cell_value is not None:
            header_cell.font = header_style
            header_cell.fill = fill_column_header

for row in ws.iter_rows(min_row=2):
    not_null = 0
    for cell in row:
        cell_value = cell.value
        if cell_value is not None:
            not_null += 1
            if not_null == 1:
                cell.fill = fill_first_column_values
            else:
                cell.fill = fill_other_columns
            

wb.save('pub_analise.xlsx')
```

#### Pergunta 4: Quais sao os produtos que normalmente sao vendidos juntos?

```python
df = dataframe[dataframe['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df
```

```python
from itertools import combinations
from collections import Counter

view_items = pd.DataFrame()
count = Counter()
product1 = []
product2 = []
count_list = []

for row in df['Grouped']:
    product_list = row.split(',')
    count.update(Counter(combinations(product_list, 2)))

for key, value in count.most_common(10):
    products_list = list(key)
    product1.append(products_list[0])
    product2.append(products_list[1])
    count_list.append(value)

view_items['Top'] = range(1,11)
view_items['Product1'] = product1
view_items['Product2'] = product2
view_items['Count'] = count_list

view_items
```

```python
with pd.ExcelWriter('top_sales.xlsx', engine='xlsxwriter') as writer:
    view_items.to_excel(writer, sheet_name='Produtos_Conjunto', index=False,startrow=2, startcol=7)
    worksheet = writer.sheets['Produtos_Conjunto']
    worksheet.title = 'Top Vendas 2 Produtos'


wb = load_workbook('top_sales.xlsx')
ws = wb['Produtos_Conjunto']

header_style = Font(color='FFFFFF', bold=True)
fill_column_header = PatternFill(start_color='1F497D', end_color='1F497D', fill_type='solid')
fill_first_column_values = PatternFill(start_color='4F81BD',end_color='4F81BD',fill_type='solid')
fill_other_columns =  PatternFill(start_color='D9E1F2',end_color='D9E1F2',fill_type='solid')

for row in ws.iter_rows(min_row=3, max_row=3):
    for i, header_cell in enumerate(row, start=3):
        cell_value = header_cell.value
        if cell_value is not None:
            header_cell.font = header_style
            header_cell.fill = fill_column_header

for row in ws.iter_rows(min_row=4):
    not_null = 0
    for cell in row:
        cell_value = cell.value
        if cell_value is not None:
            not_null += 1
            if not_null == 1:
                cell.fill = fill_first_column_values
            else:
                cell.fill = fill_other_columns
            
wb.save('top_sales.xlsx')
```

#### Pergunta 5: Qual foi o produto mais vendido?

```python

most_sold = pd.DataFrame()

most_sold['Product'] = dataframe['Product']
most_sold['Quantity Ordered'] = dataframe['Quantity Ordered']
most_sold['Price Each'] = dataframe['Price Each']

product_qnt = most_sold.groupby('Product').sum()['Quantity Ordered']
```

```python
product_table = pd.DataFrame()
product_list = [product for product, pd in most_sold.groupby('Product')]
quantity_ordered = []

for v in product_qnt:
    quantity_ordered.append(v)

product_table['Produto'] = product_list
product_table['Quantidade Encomendada'] = quantity_ordered

product_table
```

```python
fig, ax = plt.subplots()
product_table.plot(x='Produto', y='Quantidade Encomendada', ax=ax, xlabel='Numero Encomendas', ylabel='Nome Produto')

produtos = range(len(product_list))
ax.grid()
ax.set_xticks(produtos)
ax.set_xticklabels([produto for produto in product_list],rotation=90, ha='right')

image_stream_path ='quantidade_produtos.png'
plt.savefig(image_stream_path, format='png', bbox_inches='tight')
plt.close(fig)

with pd.ExcelWriter('quantidade_produtos_encomendas.xlsx', engine='xlsxwriter') as writer:
    product_table.to_excel(writer, sheet_name='Produtos_Encomendas',index=False, startrow=2, startcol=7)
    
    worksheet = writer.sheets['Produtos_Encomendas']
    worksheet.title = 'Quantidade de Produtos encomendados'

    worksheet.insert_image('K10', image_stream_path, {'x_scale':0.5, 'y_scale':0.5})

```
